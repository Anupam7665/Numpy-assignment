Q1. How can you create a Bokeh plot using Python code?

```python
from bokeh.io import output_notebook
import bokeh.plotting
from bokeh.plotting import figure , output_file , show
from bokeh.sampledata.iris import flowers
output_notebook()
```

```python
flowers.head()
```

```python
p = figure(title = "test flower")
p.xaxis.axis_label = "petal length"
p.yaxis.axis_label = "petal width"
p.circle(flowers['petal_length'] , flowers['petal_width'])
show(p)
```

Q2. What are glyphs in Bokeh, and how can you add them to a Bokeh plot? Explain with an example.

```python
x = [1, 2, 3, 4, 5]
y = [6, 7, 2, 4, 5]

p = figure(title='Scatter Plot Example', x_axis_label='X', y_axis_label='Y')
p.circle(x=x, y=y, size=10, color='navy')

show(p)
```

Q3. How can you customize the appearance of a Bokeh plot, including the axes, title, and legend?

```python
p = figure(x_range=(0, 10),
           y_range=(0, 20),
           x_axis_label='X-axis',
           y_axis_label='Y-axis',
           title='Scatter Plot',
           )
p.circle([1, 2, 3, 4, 5],
         [6, 7, 2, 4, 5],
         size=10,
         color='navy',
         legend_label='Data Points')
p.legend.location = 'top_left'

show(p)
```

Q5. How can you embed a Bokeh plot into a web page or dashboard using Flask or Django?

```python
from flask import Flask, render_template
from bokeh.embed import server_document
from bokeh.server.server import Server

# Define the Bokeh app
def bokeh_app(doc):
    p = figure(x_range=(0, 10),
           y_range=(0, 20),
           x_axis_label='X-axis',
           y_axis_label='Y-axis',
           title='Scatter Plot',
           )
    p.circle([1, 2, 3, 4, 5],
            [6, 7, 2, 4, 5],
            size=10,
            color='navy',
            legend_label='Data Points')
    p.legend.location = 'top_left'

# Initialize the Flask app
app = Flask(__name__)

# Register a route that serves the Bokeh app
@app.route('/bokeh')
def bokeh_route():
    script = server_document('http://localhost:5006/myapp')
    return render_template('bokeh.html', script=script)

# Define the Bokeh server and start it
bokeh_server = Server({'/myapp': bokeh_app})
bokeh_server.start()

# Start the Flask app
if __name__ == '__main__':
    app.run(debug=True)
```
