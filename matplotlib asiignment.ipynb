### Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data plot a scatter plot.

Scatter plot is a type of graph that show the relationship between two-axis by using the dot. Each dot is represented a x-axis and y-axis . By using 
scatter you can use a scatter plot to see how the height and weight of a group of people are related . 

```python
import numpy as np
import matplotlib.pyplot as plt 

np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.scatter(x,y)
```

### Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.

The subplot() function is used to create multiple plot in single figure. By specifying the no of rows and columns . It providing reasonable control over the how individual plots are created 

```python
import numpy as np
import matplotlib.pyplot as plt 
```

```python
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])


x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])


x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])


x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])


fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(10, 10))


ax1.plot(x1, y1 , c = "red")
ax1.set_title('Line 1')


ax2.plot(x2, y2 , c = "blue")
ax2.set_title('Line 2')


ax3.plot(x3, y3 , c = "green")
ax3.set_title('Line 3')


ax4.plot(x4, y4 , c = "orange")
ax4.set_title('Line 4')


plt.show()
```

### |Q5: What is a box plot? Why is it used? Using the following data plot a box plot.

Box plot is a graphical representation of the distribution of data based on five number summary: **** minimum, first quartile, median, third quartile, and maximum. *
It is commonly used for identifying outliers and helps to understanding the spread and skewness of the data.

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
plt.boxplot([box1,box2])
plt.show()
```
